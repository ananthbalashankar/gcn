```python
import pandas as pd
```

```python

d = pd.read_csv('./InSilicoSize100/InSilicoSize100-Ecoli1-trajectories.tsv', delimiter='\t')
all_series = {}
for c in d.columns.values:
    if 'Time' in c:
        continue
    for i in range(46):
        all_series['{0}_{1}'.format(c, i)] = d[c].iloc[21*i: 21*i + 21].values
        
all_series = pd.DataFrame(data=all_series)
all_series.to_pickle('./ecoli1-ts')
```

```python
import numpy as np
links = pd.read_pickle('dream3-ecoli1-ts_merged')
edges = {}
for k, v in links.iteritems():
    x = k[1]
    for y, c in v[3].iteritems():
        if (y,x) not in edges:
            edges[(y,x)] = []
        edges[(y,x)].append(c)
        
for k, v in edges.iteritems():
    edges[k] = np.mean(v)

links
```
